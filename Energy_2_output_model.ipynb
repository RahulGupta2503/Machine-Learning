```
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Activation, Input
from tensorflow.keras import metrics
from sklearn.metrics import r2_score
#from tensorflow.python.keras.utils.vis_utils import plot_model
from tensorflow.keras.utils import plot_model
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
```

```
df = pd.read_excel("/content/drive/MyDrive/Cases/Energy Efficiency/ENB2012_data.xlsx")
df.head()
```

```
X = df.iloc[:,:8]
y = df.iloc[:,8:]
print(X.head())
```

```
print(y.head())
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=2022)
X_train.shape
```

```
scalerX = MinMaxScaler()
scalerY = MinMaxScaler()
y_trn_scaled = scalerY.fit_transform(y_train)    
X_trn_scaled = scalerX.fit_transform(X_train) 
y_tst_scaled = scalerY.transform(y_test)  
X_tst_scaled = scalerX.transform(X_test)      
```

```
y1 = y_trn_scaled[:,0]
y1[:5]
```

```
y2 = y_trn_scaled[:,1]
y2[:5]
```

```
#tf.random.set_seed(2022)
input_layer = Input(shape = X_train.shape[1])
dense_1 = tf.keras.layers.Dense(7, activation='relu',name="Dense_1")(input_layer)
dense_2 = tf.keras.layers.Dense(5, activation='relu', name = "Dense_2")(dense_1)
dense_3 = tf.keras.layers.Dense(4, activation='relu', name = "Dense_3")(dense_2)

y1_output = tf.keras.layers.Dense(1, activation='relu', name = "y1_output")(dense_3)
dense_4 = tf.keras.layers.Dense(4, activation='relu', name = "Dense_4")(dense_2)
y2_output = tf.keras.layers.Dense(1, activation='relu', name = "y2_output")(dense_4)
model = Model(inputs = input_layer, outputs = [y1_output,y2_output])
print(model.summary())
model.compile(optimizer='adam', loss={'y1_output':'mse','y2_output':'mse'},
                                metrics={'y1_output':metrics.RootMeanSquaredError(),
                                         'y2_output':metrics.RootMeanSquaredError()})
```

```
plot_model(model, show_layer_activations=True, show_shapes=True)
```

```
from tensorflow.keras.callbacks import ReduceLROnPlateau
reduce_lr = ReduceLROnPlateau(monitor='val_y2_output_root_mean_squared_error', factor=0.2,  patience=5, min_lr=0.0005)
history = model.fit(X_trn_scaled,y_trn_scaled,validation_data=(X_tst_scaled,y_tst_scaled),callbacks=[reduce_lr,],verbose=2,epochs=500)
```

```
plt.plot(history.history['y2_output_root_mean_squared_error'])
plt.plot(history.history['val_y2_output_root_mean_squared_error'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```
plt.plot(history.history['y1_output_root_mean_squared_error'])
plt.plot(history.history['val_y1_output_root_mean_squared_error'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

```
y_pred = model.predict(X_tst_scaled)
y_pred = np.array(y_pred).reshape(y_tst_scaled.shape[0],2)
y_pred[-5:]
```

```
y_pred_orig = scalerY.inverse_transform(y_pred)
y_pred_orig[-5:]
```
